```python
def search_articles_by_category(category):
    query = {
        "query": {
            "match": {
                "predicted_category": category
            }
        }
    }
    
    results = es.search(index=INDEX_NAME, body=query)
    return [hit["_source"] for hit in results["hits"]["hits"]]

# Example: Get all "Tech" news articles
tech_articles = search_articles_by_category("Tech")

for article in tech_articles:
    print(f"{article['headline']} - {article['description']}")

```
